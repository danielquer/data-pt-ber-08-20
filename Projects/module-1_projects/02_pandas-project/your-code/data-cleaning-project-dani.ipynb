```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../attacks.csv',encoding="latin")
df.head()
```

```python
# Making dates in 'Case Number' workable
df['Case Number'] = df['Case Number'].apply(lambda x: str(x).replace('.','-'))
```

```python
# Identifying which columns have a number of nulls
df.isnull().sum() / len(df)
```

```python
# Dropping unnecessary columns full of nulls
df.drop(['Unnamed: 22','Unnamed: 23'],axis=1,inplace=True)
df
```

```python
# Renaming 'Case Number' column to Date
df.rename(columns={'Case Number':'Date'},inplace=True)
df
```

```python
# Truncating Year to the integer
df['Year'].fillna(0, inplace = True)
df['Year'] = df['Year'].apply(lambda x: int(x))
```

```python
# Removing whitespaces in column names
df.columns = map(lambda x: x.strip(),list(df.columns))
```

```python
# Truncating 'original order' to the integer
df['original order'].fillna(0, inplace = True)
df['original order'] = df['original order'].apply(lambda x: int(x))
```

```python
# Correcting values in Fatal (Y/N)
set(df['Fatal (Y/N)'])
```

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].apply(lambda x: str(x).replace('M','N').strip())
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].apply(lambda x: str(x).replace('2017','Y').strip())
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].apply(lambda x: str(x).replace('y','Y'))
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].apply(lambda x: str(x).replace('UNKNOWN',''))
```
