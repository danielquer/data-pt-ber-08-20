```python
import pandas as pd
import numpy as np
```

```python
import ast
ta = pd.read_csv('../data/ta_restaurants_curated.csv')
ta['Cuisine Style'] = ta['Cuisine Style'].fillna('[]')
ta['Cuisine Style'] = ta['Cuisine Style'].apply(lambda x: ast.literal_eval(x))
ta.head()
```

```python
ci = pd.read_excel('../data/PopulatinandArea.xlsx')
ci.head()
```

```python
df = pd.merge(ta,ci,on='City',how='left')
df
```

```python
df['Cuisine Style'].head(1)
```

```python
df['Number of Styles'] = df['Cuisine Style'].str.len()
```

```python
df['Price Range'].value_counts()
```

```python
df['Price Rank'] = np.where(df['Price Range'] == '$',1,np.where(df['Price Range'] == '$$$$',3,2))
df.head()
```

```python
df_exp = df.explode('Cuisine Style')
df_exp.head()
```

```python
final = df[['Name','City','Country','Cuisine Style','Ranking','Rating','Price Range','Price Rank','Number of Reviews','Number of Styles','Population']]

final.head()
```

```python
final_exp = df_exp[['Name','City','Country','Cuisine Style','Ranking','Rating','Price Range','Price Rank','Number of Reviews','Number of Styles','Population']]
final_exp.head()
```

```python
final.to_csv('../data/tripadvisor_restaurant_ratings.csv')
final_exp.to_csv('../data/tripadvisor_restaurant_ratings_exploded.csv')
```
