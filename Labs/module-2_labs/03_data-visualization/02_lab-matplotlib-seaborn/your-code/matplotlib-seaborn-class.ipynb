```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_palette('husl')
```

```python
help(sns.set_palette)
```

### Seaborn (https://seaborn.pydata.org/tutorial.html)

- Streamlines the amount of code needed to write to create matplotlib visualizations
- Provides specialized support for using categorical variables
- Has functions to examine relationships between multiple variables
- Provides convenient views onto the overall structure of complex datasets

```python
df = pd.DataFrame()
df['x'] = np.arange(0,10,1)
df['y'] = df['x'] ** 2

# Remember - how to plot with matplotlib

plt.plot(df['x'],df['y'])
plt.show()
```

```python
# Or with this

df.plot('x','y')
plt.show()
```

```python
# With Seaborn:
sns.lineplot(df['x'],df['y'])
```

```python
# lmplot: plotting the data together with a regression line.
sns.lmplot('x','y',data=df)
```

## Using real data: vehicles.csv

```python
data = pd.read_csv('vehicles.csv')
data.head()
```

## Boxplot

### In Matplotlib

```python
data.boxplot(column = 'Combined MPG')
```

```python
print(data['Combined MPG'].mean())
```

### In Seaborn

```python
sns.boxplot(x = 'Combined MPG', data = data)
```

```python
sns.boxplot(x = 'Combined MPG', y = 'Highway MPG', data = data)
```

## Violinplot

In Seaborn only. Matplotlib doesn't have it.

```python
sns.violinplot('Combined MPG', data = data)
```

```python
sns.violinplot(x = 'Combined MPG', y = 'Highway MPG', data = data)
```

## Histogram

### In Matplotlib

```python
plt.hist(data['Combined MPG'],bins=20)
```

### In Seaborn

```python
sns.distplot(data['Combined MPG'])
```

## Bar Plot

```python
help(sns.barplot) ## for categorical data
```

```python
fig = sns.barplot(x = 'Highway MPG', y = 'Fuel Type', data = data).get_figure()
fig.savefig('barplot_highway_fueltype.png')
```

```python
sns.barplot(x = 'Highway MPG', y = 'Fuel Type', data = data, estimator = np.median)
```

```python
sns.countplot('Fuel Type', data = data)
```
