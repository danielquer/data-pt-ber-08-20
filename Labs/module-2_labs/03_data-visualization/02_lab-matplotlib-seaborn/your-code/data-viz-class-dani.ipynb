```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vehicles.csv')
```

```python
df.head()
```

```python
plt.scatter(df['City MPG'],df['Combined MPG'])
```

```python
plt.scatter(df['City MPG'],df['Highway MPG'])
```

```python
df['City MPG'].corr(df['Combined MPG'])
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(), vmin = -1, vmax = 1, cmap = 'RdYlGn', annot = True)
```

```python
df.corr(method = 'spearman')
```

```python
sns.heatmap(df.corr(method = 'spearman'), vmin = -1, vmax = 1, cmap = 'RdYlGn', annot = True)
```

##Pivot Tables

```python
help(pd.pivot_table)
```

```python
df.pivot_table(index = 'Vehicle Class', aggfunc = np.median)
```

```python
df.pivot_table(index = 'Drivetrain',aggfunc = np.median, columns='Year', values = 'City MPG', fill_value = 0)
```

```python
df.pivot_table(index =['Vehicle Class','Drivetrain'],aggfunc = np.median, values = 'City MPG', fill_value = 0)
```
