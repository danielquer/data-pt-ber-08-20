## Continous Probability Distributions

To calculate probability of continuous variables, we compute the total area under the distribution curve of the sample space. At a given point of the curve, the area is approximately zero.
We can convert a continuous distribution to descrete distribution by using multiple bins where each bin represents the probability value of a certain range.

```python
# Probability Density Function
# P(a <= x <= b) = integral (from a to b) of derivative of function of x
```

```python
# Cumulative Distribution Function
#Â F(a) = P(X E (-Infinite, a]) = integral (from -infinite to a) of derivative of function of x
```

```python
# Continuous Uniform Distribution. Example of one
import numpy as np
from matplotlib import pyplot as plt
np.random.random()
```

```python
from scipy.stats import uniform
```

```python
cuni = uniform()
X = cuni.rvs(size=100)

plt.hist(X)
```

```python
numrange = np.linspace(-0.1,1.1,100)
```

```python
plt.plot(numrange,cuni.pdf(numrange))
```

```python
plt.plot(numrange,cuni.cdf(numrange))
```

```python
cuni = uniform(0,10)
numrange = np.linspace(-1,11,100)
plt.plot(numrange,cuni.pdf(numrange))
```

##Â Normal distribution

```python
from scipy.stats import norm
mean = 0
std = 1
normal = norm(mean,std)
```

```python
numrange = np.linspace(-4,4,100)
plt.plot(numrange,normal.pdf(numrange))
```

```python
numrange = np.linspace(-4,4,100)
plt.plot(numrange,normal.cdf(numrange))
```

```python
X = normal.rvs(size = 10000)
plt.hist(X,bins=100)
plt.show()
```

```python
#Â Skewness is a measure of symmetry. 
#Â left skewed - median < mean. Some outliers on the beginning are dragging the mean down compared to the median. E.g. Age of deaths in western societies.
#Â right skewed - mean > median. Some outliers on the end are dragging the mean up compared to the median. E.g. Income, poisson distributions
```

```python
#Â Kurtosis is a measure of how many outliers you have in your distribution
# - leptokurtic - less tails, less outliers
#Â - mesokurtic - a normal distribution
# - platykurtic - fat tails, more outliers
```
