### What is an API?

APIs act as intermediaries between two applications, between a mobile app and a server, between a data provider and a server...

They are building blocks for developers, they regulate data access, they are a communication point, for example embedding content, or an OAuth standard.

There are many types of API Protocols, but the most common one is the REST API
* REST API: type of API that will give data on a JSON format. You can use requests such as DELETE, GET, PUT, POST, and create CUSTOM parameters.

### Nested Data

### GitHub API
https://developer.github.com/v3/activity/feeds

```python
feedsurl = "https://api.github.com/feeds"
```

```python
response = requests.get(feedsurl)
gitfeeds = response.json()
pd.DataFrame(gitfeeds) ### showing a df with 4 columns, 3 rows. One of the columns contain nested data.
```

```python
## Seeing one more emample of the above, now clearer: one column with nested data. 
caturl = "https://cat-fact.herokuapp.com/facts"
response = requests.get(caturl)
results = response.json()
pd.DataFrame(results) 
```

```python
## we can actually see how many columns a dataframe from a json file will contain by running the following:
results.keys()
```

```python
## Instead, we only want data nested inside the key. We can call the key 'all' as if it was a dictionary.
pd.DataFrame(results['all'])
```

```python
## Alternative, we can use the pandas function pd.json_normalize()
pd.json_normalize(results['all'])
```
