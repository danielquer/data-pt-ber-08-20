### Writing a function to make a soup from a URL

```python
def make_soup(url):
    html = requests.get(url).content
    soup = BeautifulSoup(html,'lxml')
    return soup
```

```python
make_soup("https://elpais.com/deportes/2020-10-01/sorteo-de-la-fase-de-grupos-de-la-champions-league-2020-en-directo.html")
```

### Scraping a Wikipedia article

```python
wikisoup = make_soup("https://en.wikipedia.org/wiki/List_of_police_dog_breeds")
# find all elements in soup that belong to a list
wikisoup.find_all('li')
# take all elements in this find_all that start with an upper case letter
[i.text.replace('[A-Z]','') for i in wikisoup.find_all('li') if re.match('[A-Z]',i.text)][0:17]
```

```python
tags = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'p', 'li']
wikitextlist = [element.text for element in wikisoup.find_all(tags)]
wikitextlist
```
