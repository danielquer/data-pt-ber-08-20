### Challenge 1: Forks

#### Obtain the full list of forks created from the main lab repo via the Github API.


```python
# Create variables which contain your user name and your token
GITHUB_USERNAME = ''
GITHUB_ACCESS_TOKEN = '' 
```

```python
# Retrieve a json file with information on all of the forks. You will need an "auth" argument here for authentication!
base = 'https://api.github.com/repos/ironhack-datalabs/data-pt-ber-08-20'
forksurl = base + '/forks'
response = requests.get(forksurl,auth=(GITHUB_USERNAME,GITHUB_ACCESS_TOKEN))
results = response.json()
results
```

```python
# Normalize semi-structured JSON data into a dataframe. 
unnormalized_df = pd.DataFrame(results)
df = pd.json_normalize(unnormalized_df['owner'])
df
```

#### Extract a list of all the usernames of people who have forked this repo. 

```python
# Your code:
usernames = [i for i in df['login']]
usernames
```

### Challenge 2: Count Commits

Obtain a dataframe which contains all of the commits made. From this dataframe, extract a list of all the commit messages. Also, count how many commits have been made.

```python
# Your code:
commitsurl = base + '/commits'
response = requests.get(commitsurl,auth=(GITHUB_USERNAME,GITHUB_ACCESS_TOKEN))
results = response.json()
results
```

```python
commits_list = [i for i in pd.json_normalize(results)['commit.message']]
commits_list
```

```python
count = pd.json_normalize(results)['commit.message'].count()
print(f'There have been {count} commits.')
```

#### Bonus Challenge: Obtain all of the commits that have been made in the past week (specify the date using a 'since' argument).

```python
# Your code:
param_dict={'since':'2020-10-03'}
commitsurl = base + '/commits'
response = requests.get(commitsurl,auth=(GITHUB_USERNAME,GITHUB_ACCESS_TOKEN),params=param_dict)
results = response.json()
results
```

```python
count = pd.json_normalize(results)['commit.message'].count()
print(f'There have been {count} commits in the past week.')
```
