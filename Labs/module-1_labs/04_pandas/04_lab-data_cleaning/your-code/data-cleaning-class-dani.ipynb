```python
df.columns = map(lambda x: x.upper(),df.columns)
df.columns
```

Investigating and replacing null values

```python
# Check what values are null
df.isnull()
# Check how many values are null for each column
df.isnull().sum()
checking_null_cols = df.isnull().sum() / len(df)
checking_null_cols
```

```python
invalid_columns = checking_null_cols[checking_null_cols > 0.33]
list_columns_to_ditch = list(invalid_columns.index)
list(checking_null_cols[checking_null_cols > 0.33].index)
```

```python
# Check the count of non-missing values for each column
df.count()
```

```python
# Dropping columns that have more missing values than desired. 
df.drop(list(checking_null_cols[checking_null_cols > 0.33].index), axis = 1)
```

```python
# Fuel type: 'Electricity' means that they will have no cylinders and no displacement. We'll replace them by 0.
df['CYLINDERS'].isnull().sum()
```

```python
df['CYLINDERS'].fillna(0, inplace = True)
```

```python
df['CYLINDERS'].isnull().sum()
```
