#### Import all the libraries that are necessary.

```python
# import libraries here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Define your own data. We would like to have a variable callex which contains a list of numbers from 0 to 100. Furthermore, we would like a variables called y which is twice x, and a variable called z which is the square of x.

```python
# your code here
x = [i for i in range(0,101)]
y = [2*i for i in x]
z = [i ** 2 for i in x]
print(x)
print(y)
print(z)
```

#### Plot (x,y) and (x,z) on the axes in a line plot.

#### There are 2 ways of doing this - the way we have seen in class, and also using a subplot. Please research and try both!

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here - 1st way
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
# your code here - 2nd way
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)

plt.show()
```

#### Augmenting your previous code, label your axes.

```python
# your code here
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("X axis")
plt.ylabel("Functions of X")
plt.legend(['y','z'])
plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
import math
expx = [math.exp(i) for i in x]
plt.plot(x,expx)
plt.plot(x,z)
plt.show()
```

```python
figure, axs = plt.subplots(1,2)
ax = axs[0]
ax.plot(x,expx)
ax.plot(x,z)
ax.set_yscale('log')
plt.show()
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
plt.bar?
```

```python
# your code here
grouped = df[['City MPG','Fuel Type']].groupby(['Fuel Type']).sum().reset_index()

plt.bar(grouped['Fuel Type'],height=grouped['City MPG'])
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# your code here
grouped_barrels = df[['Model','Fuel Barrels/Year']].groupby(['Fuel Barrels/Year']).count().reset_index()

grouped_barrels
```
